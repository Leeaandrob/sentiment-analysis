```python
import requests
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# get comments page posts from CCV - PRINCIPAL
data = requests.get('http://localhost:8080/api/comments/?page_name=cocacolabr').json()
data = data['data']

post_comments = [[comment['message'] for comment in post['comments']['data']] for post in data]
comments = post_comments[:3][0]

# instance my countvector
vect = CountVectorizer()

vect.fit(comments)

# examine the fitted vocabulary
vect.get_feature_names()

# transform training data into a 'document-term matrix'
post_comments_dtm = vect.transform(comments)
```

```python
post_comments_dtm
```

```python
# convert sparse matrix to a dense matrix
post_comments_dtm.toarray()
```

```python
# examine the vocabulary and document-term matrix together
pd.DataFrame(post_comments_dtm.toarray(), columns=vect.get_feature_names())
```
